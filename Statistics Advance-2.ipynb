

### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

**Probability Mass Function (PMF):**
- The PMF is a function that gives the probability that a discrete random variable is exactly equal to some value. It is used for discrete random variables.
- **Example:** Suppose we roll a fair six-sided die. The PMF of this experiment assigns a probability to each outcome (1, 2, 3, 4, 5, 6), which is \( \frac{1}{6} \) for each value.

**Probability Density Function (PDF):**
- The PDF is a function that describes the likelihood of a continuous random variable to take on a particular value. The probability of the variable falling within a particular range is given by the area under the curve of the PDF within that range.
- **Example:** For a standard normal distribution (mean = 0, standard deviation = 1), the PDF is given by the function \( f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \).

### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why is CDF used?

**Cumulative Density Function (CDF):**
- The CDF of a random variable \( X \) is a function that gives the probability that \( X \) will take a value less than or equal to \( x \). It is used for both discrete and continuous random variables.
- **Example:** If \( X \) is the outcome of rolling a fair six-sided die, the CDF at \( x = 3 \) is the probability that the outcome is less than or equal to 3, which is \( P(X \leq 3) = \frac{3}{6} = 0.5 \).

**Why CDF is used:**
- The CDF provides a complete description of the probability distribution of a random variable.
- It is useful for calculating probabilities over intervals and for understanding the distribution's behavior.

### Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

**Examples of Normal Distribution:**
- Heights of people in a population.
- Test scores in a large exam.
- Measurement errors in scientific experiments.
- Daily returns of a stock in finance.

**Parameters and Shape:**
- **Mean (\(\mu\))**: Determines the center of the distribution.
- **Standard Deviation (\(\sigma\))**: Determines the spread or width of the distribution.
  - A larger \(\sigma\) results in a wider and flatter distribution.
  - A smaller \(\sigma\) results in a narrower and taller distribution.

### Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

**Importance of Normal Distribution:**
- Many statistical methods and tests assume normality due to its mathematical properties.
- The Central Limit Theorem states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population's distribution.
- It is widely used in inferential statistics to make predictions and decisions based on data.

**Real-life Examples:**
- Heights of adult men and women.
- IQ scores.
- Blood pressure measurements.
- Measurement errors in physical experiments.
- Stock returns in financial markets.

### Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

**Bernoulli Distribution:**
- A Bernoulli distribution is a discrete probability distribution for a random variable which takes the value 1 with probability \( p \) and the value 0 with probability \( 1-p \).
- **Example:** Flipping a coin, where 1 represents heads and 0 represents tails, with \( p = 0.5 \).

**Difference between Bernoulli Distribution and Binomial Distribution:**
- **Bernoulli Distribution:** Represents a single trial with two possible outcomes (success or failure).
- **Binomial Distribution:** Represents the number of successes in a fixed number of independent Bernoulli trials with the same probability of success.

### Q6: Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

Given:
- Mean (\(\mu\)) = 50
- Standard Deviation (\(\sigma\)) = 10
- Value (\(X\)) = 60

**Calculation:**
1. Convert the value to a z-score:
   \[
   z = \frac{X - \mu}{\sigma} = \frac{60 - 50}{10} = 1
   \]

2. Use the standard normal distribution to find the probability:
   \[
   P(Z > 1)
   \]

3. Look up the z-score in standard normal distribution tables or use a calculator to find:
   \[
   P(Z > 1) \approx 0.1587
   \]

So, the probability that a randomly selected observation will be greater than 60 is approximately 0.1587 or 15.87%.

### Q7: Explain Uniform Distribution with an example.

**Uniform Distribution:**
- A uniform distribution is a type of probability distribution in which all outcomes are equally likely. Each value in a finite set of possible values has an equal probability.

**Example:**
- Rolling a fair six-sided die. Each outcome (1 through 6) has an equal probability of \( \frac{1}{6} \).

### Q8: What is the z-score? State the importance of the z-score.

**Z-score:**
- The z-score is a measure of how many standard deviations an element is from the mean of the distribution. It is calculated as:
  \[
  z = \frac{X - \mu}{\sigma}
  \]
  where \( X \) is the value, \( \mu \) is the mean, and \( \sigma \) is the standard deviation.

**Importance of the Z-score:**
- Standardizes different datasets for comparison.
- Identifies outliers.
- Calculates probabilities and percentiles in a normal distribution.

### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

**Central Limit Theorem (CLT):**
- The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes larger, regardless of the population's distribution.

**Significance of CLT:**
- Allows us to use normal distribution techniques for inference even if the data is not normally distributed, given a sufficiently large sample size.
- Justifies the use of sample means for hypothesis testing and confidence intervals.

### Q10: State the assumptions of the Central Limit Theorem.

**Assumptions of CLT:**
1. The samples must be independent of each other.
2. The sample size should be sufficiently large (commonly \( n > 30 \)).
3. The population from which the samples are drawn should have a finite variance.
